```python
#Gr√°fico BTCxETH em 2022 com Python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
criptos = yf.download(['BTC-USD', 'ETH-USD'], start = '2021-12-31', end = '2022-12-31')['Adj Close']
criptos.plot(kind = 'line', secondary_y = 'ETH-USD')
plt.title("BTC x ETH", color = 'w')
plt.show()
```
